```python
#Access results URL for runner #6541 (me)
x = urllib2.urlopen("http://www.mtecresults.com/runner/show?race=2787&email=1&rid=6541")
page = x.read()
#Function to extract name and gender from HTML format: itemprop="name" content="FirstName LastName" (inputs: HTML and item type)
def get_info(results, category):
    category = 'itemprop="' + category
    item = results.find(category)
    item = results.find('content=', item)
    start = results.find('"', item)
    end = results.find('"', start + 1)
    return results[start+1:end]
#Function to extract age, city, and state from HTML format: itemprop="age">XX<" (inputs: HTML and item type)
def get_info2(results, category):
    category = 'itemprop="' + category
    item = results.find(category)
    start = results.find('>', item)
    end = results.find('<', start)
    return results[start+1:end]
#Function to extract race results (times/rankings) from format: <th class='runnerresultsfull'>XX:XX:XX</th> (inputs: HTML, row label)
def get_times(results, category):
    times = []
    category = "<th class='runnerresultsfull'>" + category + "</th>"
    item = results.find(category)
    #Loop through each of the 7 columns in the results table and extract the values into the times array:
    for i in xrange(7):
        item = results.find("<th class='runnerresultsfull", item + 1)
        start = results.find(">", item)
        end = results.find("<", start)
        time = results[start+1:end]
        times.append(time)
    return times
#Extract all data and print for evaluation of relevant values:
name = get_info(page, 'name')
gender = get_info(page, 'gender')
age = get_info2(page, 'age')
city = get_info2(page, 'addressLocality')
state = get_info2(page, 'addressRegion')
headers = get_times(page, 'Location')
overall = get_times(page, 'Time')
chip = get_times(page, 'ChipStart')
division = page[page.find("=", page.find('/race/show/2787?division='))+1:page.find('"', page.find('/race/show/2787?division='))]
    
print name, gender, age, city, state, division
print headers
print overall
print chip

x.close()
```

I tried splitting the data set into training and testing data and then running a linear regression algorithm to see how well I could predict runners' finish times from thier bib numbers (assigned sequentially by estimated finish times at the time of registration).  While there was a clear linear association between these two variables, the model did okay but not as good as I had hoped.  The average squared residual was 865018 seconds^2^.  This suggested that many runners' estimations were inaccurate, so using bib numbers to organize waves, while an okay solution did not produce ideal results.

```python
results_lin = results_all[['Bib']]
X = results_lin.as_matrix()
Y = results_all['Time_Sec']
x_train, x_test, y_train, y_test = sklearn.cross_validation.train_test_split(X, Y, test_size=0.33, random_state=99)
lm = sklearn.linear_model.LinearRegression()

results_lin = results_all[['Bib']]
lm.fit(x_train, y_train)

avgsqres = np.mean((results_all['Time_Sec'] - lm.predict(results_lin))**2)
print 'Coefficients: ' + str(lm.coef_)
print 'Average Square Residuals: ' + str(round((avgsqres), 0)) + '(seconds^2)'

fig = plt.figure()
fig.set_size_inches(18, 6)
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.scatter(x_test, y_test, alpha=0.5)
ax1.plot(x_test, lm.predict(x_test), color='red')
ax1.set_title('Linear Regression for Race Finish Times From Bib Number')
ax1.set_xlabel('Bib Number')
ax1.set_ylabel('Finish Time (seconds)')
ax1.set_xlim(0, 20000)
ax2.scatter(y_test, lm.predict(x_test), alpha=0.5)
ax2.set_title('Linear Regression Prediction vs. Actual')
ax2.set_xlabel('Actual Finish Time (seconds)')
ax2.set_ylabel('Predicted Finish Times (seconds)')
ax2.set_xlim(2000, 8000)
ax2.set_ylim(2000, 8000)
plt.show()
```

Addressing the issue of determining how to divide up the waves for the race further, I decided to look at the comparison between runners' assigned waves and the actual waves they started in.  Assigned waves were not enforced but runners were instructed that they could only start in a later wave than the wave to which they were assigned, not an earlier one.  In order to look at this I examined the data to come up with the actual start times for each wave.  I discovered that there were actually 11 waves even though there were only 10 assigned waves.  It appears that an additional wave was added on the end (the last start time was supposed to be 9:30 but wave 11 started at 9:40).  This may have been due to logistical issues, as runners had to be bussed from satellite parking lots.  It highlights an issue that the race organizers may want to look into for next year's logistical planning.

```python
#Create datetime objects for the actual start times of each wave (based on examination of the data)
wave2 = dt.datetime(2014, 11, 9, 7, 26)
wave3 = dt.datetime(2014, 11, 9, 7, 40)
wave4 = dt.datetime(2014, 11, 9, 7, 56)
wave5 = dt.datetime(2014, 11, 9, 8, 10)
wave6 = dt.datetime(2014, 11, 9, 8, 38)
wave7 = dt.datetime(2014, 11, 9, 8, 48)
wave8 = dt.datetime(2014, 11, 9, 8, 58)
wave9 = dt.datetime(2014, 11, 9, 9, 13)
wave10 = dt.datetime(2014, 11, 9, 9, 30)
wave11 = dt.datetime(2014, 11, 9, 9, 40)
#Save actual start times to a column in the data frame
results_all['Start_Wave'] = 11
results_all['Start_Wave'][results_all['Start_Time'] < wave11] = 10
results_all['Start_Wave'][results_all['Start_Time'] < wave10] = 9
results_all['Start_Wave'][results_all['Start_Time'] < wave9] = 8
results_all['Start_Wave'][results_all['Start_Time'] < wave8] = 7
results_all['Start_Wave'][results_all['Start_Time'] < wave7] = 6
results_all['Start_Wave'][results_all['Start_Time'] < wave6] = 5
results_all['Start_Wave'][results_all['Start_Time'] < wave5] = 4
results_all['Start_Wave'][results_all['Start_Time'] < wave4] = 3
results_all['Start_Wave'][results_all['Start_Time'] < wave3] = 2
results_all['Start_Wave'][results_all['Start_Time'] < wave2] = 1
results_all.to_csv('results_all1.csv')
```

Based on these results I would suggest for next year that the guidance for bus boarding times for earlier waves be adjusted to 30-45 minutes prior to the race instead of 1 hour, while later waves could remain at 1 hour.

Earlier I looked at the distribution of finish times, which was a skewed distribution.  To further examine gender differences I broke this down by gender below.  Both the male and female runner groups had the same shape skewed distribution with a very long tail reaching out to about the 4:00:00 mark (approximately a 38.7 minute mile pace).  The median finish times (1:10:01 for females and 1:02:08 for males) were higher than the national averages according to Running USA (1:04:47 for females and 56:00 for males).  With such a large sample size of almost 15000 finishers, I would expect these numbers to be much closer to the national averages.  Some possible explanations are that since the annual Bay Bridge walk was cancelled several years ago, many walkers signed up, skewing the times upward since walkers typically do not participate in 10K races.  The race across the bridge was very scenic, and many runners stopped to enjoy the views and take photos which likely impacted times.  Also, the nature of the course is slightly more difficult than a typical 10K course as the first half of the race took place on an incline.  This is important to take into account for planning for next year, as the cost of closing the bridge is extremely high not only in terms of causing traffic, but also in terms of required security and lost tolls due to people avoiding travel during that time frame.  Some possible mitigations for this issue would be to have a pace car to pick up lagging runners, or start a small number of the slowest runners in the first wave instead of the last wave so that they finish during the regular course of the race.  The actual race duration (based off of the minimum start time and maximum finish time) was about 5 hours (07:00:01 to 12:06:47) which was about 2 hours and 26 minutes after the start of the last wave.  Considering that the slowest runners were supposed to be in the last wave, the outliers must have started in much earlier waves.

```python
#Load 2006 Bay Bridge 10K results
x = urllib2.urlopen("http://annapolisstriders.org/GBBROVAllResults06.htm")
page2 = x.read()
#Set initial page marker
marker2 = page2.find('Chiptime') 
oldresults = pd.DataFrame(columns=['Rank', 'Name', 'Gender', 'Age', 'City', 'State', 'Time', 'Time_Sec'])
#Loop through each runner and extract data
for j in xrange(2144):
    item_old=[]
    #Loop through each item
    for i in xrange(6):
        #New runners start with <td height tags
        if i == 0:
            marker = page2.find('<td height', marker2)
        #Runner info starts with <td class tags
        else:
            marker = page2.find('<td class', marker2)
        marker = page2.find('>', marker)
        marker2 = page2.find('<', marker)
        #Remove new line characters
        item1 = page2[marker+1:marker2].replace('\n', '').replace('\r', '')
        item_old.append(item1)
        
    rank = int(item_old[0].split('. ')[0].strip())
    names = item_old[0].split('. ')[1].strip()
    #Account for names with only first initial
    if len(item_old[0].split('. ')) > 2:
        names += '. ' + item_old[0].split('. ')[2].strip()
    name = names.split(' ')[0].rstrip() + ' ' + names.split(' ')[-1]
    gender = item_old[2][0].upper()  #Change all lowercase genders to uppercase
    age = int(item_old[2][1:])
    city = item_old[3].split(', ')[0]
    #Assign 'INT' state to international runners
    if len(item_old[3].split(', ')) > 1: 
        state = item_old[3].split(', ')[1]
    else:
        state = 'INT'
    time = item_old[5].split('.')[0]
    #Calculate time in seconds
    if len(time.split(":")) > 2:
        time_sec = int(time.split(":")[-3])*3600 + int(time.split(":")[-2])*60 + int(time.split(":")[-1])
    else:
        time_sec = int(time.split(":")[-2])*60 + int(time.split(":")[-1])
    #Add runner to data frame
    oldresults.loc[len(oldresults)+1]=[rank, name, gender, age, city, state, time, time_sec]
#Save to CSV    
oldresults.to_csv('oldresults.csv')
```

In order to better capture the impact of passing on wave organization, I created a new metric, the inter-wave passing rate, to measure how many runners passed/were passed by runners from different waves.  I believe this metric is more accurate in describing the impact of passing on the race because with large waves of runners not being organized in speed order there will inevitably be lots of passing within each wave. However, if we orgainze the waves optimally we shouldn't see much passing between runners starting in different waves.  To calculate this metric I looped through each runner and counted how many runners from different waves either passed or were passed by that runner.  Then I stored the results in the data frame.  The results were very similar to the overall passing rate, although lower.  The pattern was the same though - there was a sharp increase in the inter-wave passing rate at Wave 7.  One thing that may have caused this is that as the runners' average paces increases, the 15 minutes between waves becomes less significant for preventing passing because runners are on the course for longer.  Interestingly and perhaps unexpectedly, the total inter-wave passing events reached over 1 million in Wave 10.

```python
#Calculate Inter-Wave passing rates
results_all['R_Passed_By'] = 0
results_all['R_Passed'] = 0
for runner in range(0, len(results_all)):
    starttime = results_all.ix[runner, 'Start_Time']
    finishtime = results_all.ix[runner, 'Finish_Time']
    #Limit to runners in other actual start waves
    wave = results_all.ix[runner, 'Start_Wave']
    rwave = results_all[results_all['Start_Wave'] != wave]
    #Select runners who passed the current runner
    rpassedby = rwave[rwave['Start_Time'] > starttime]
    rpassedby = rpassedby[rpassedby['Finish_Time'] < finishtime]
    #Select runners who were passed by the current runner
    rpassed = rwave[rwave['Start_Time'] < starttime]
    rpassed = rpassed[rpassed['Finish_Time'] > finishtime]
    #Add values to dataframe
    results_all.ix[runner, 'R_Passed_By'] = int(len(rpassedby))
    results_all.ix[runner, 'R_Passed'] = int(len(rpassed))
```

To determine whether the running course was too crowded, I decided to model the runner density per block of the course.  I accomplished this by breaking the 10K (6.2 mile) course up into blocks of a set size (after some testing I settled on 10 foot blocks).  Then I looped through each 10 minute increment of the race and created a column in the timeblocks data frame to record every runner's block position at each time (using -1 to represent runners who already finished and -2 to represent runners who hadn't started yet).  Finally I plotted a histogram for each time interval to visualize the runner density (I only showed 3 in this notebook to avoid making it too big but I created a separate notebook with all 30 plots for 10 minute increments).  This proved to be a very effective method for modeling runner density.

```python
#Set block size in feet
blocksize = 10
#Calculate number of blocks in the race course (10k = 6.2 miles)
num_blocks = np.ceil((5280*6.2)/blocksize)
#Set racestart to equal the start time of the first runner
racestart = min(results_all['Start_Time'])
#Convert start and finish times to seconds (divide by 1 second to convert to float, then convert to int)
results_all['Start_Time_Sec'] = (results_all['Start_Time'] - racestart)/dt.timedelta(seconds=1)
results_all['Start_Time_Sec'] = results_all['Start_Time_Sec'].astype('int')
results_all['Finish_Time_Sec'] = results_all['Start_Time_Sec'] + results_all['Time_Sec']
timeblocks = results_all[['Bib', 'Start_Time_Sec', 'Finish_Time_Sec', 'Time_Sec']]
#Calculate runner block pace (in blocks/second)
timeblocks['Pace'] = num_blocks/(timeblocks['Time_Sec'].astype('float'))

#Create array of times to take measurements
intervals = []
time_counter = 0
#Loop through and create a column for each 5 minute increment of the race that records the position of all runners at that time
while time_counter < 18000:
        #Use increments of 10 minutes
        time_counter += 600
        intervals.append(time_counter)
        #Form column name as Pos_[Time]
        col_name = 'Pos_' + str(time_counter)
        #Calculate block position at the time for each runner
        timeblocks[col_name] = ((time_counter - timeblocks['Start_Time_Sec'])*(timeblocks['Pace']) + 1).astype('int')
        #Set block value of runners who haven't started yet to -2
        timeblocks[col_name][timeblocks[col_name] < 1] = -2
        #Set block value of runners who finished already to -1
        timeblocks[col_name][timeblocks[col_name] > num_blocks] = -1
```

```python
blocks_per_quarter_mile = num_blocks/(6.2*4)
#Calculate x-axis labels to show mile markers instead of block number
axseries = np.arange(0, num_blocks + blocks_per_quarter_mile, blocks_per_quarter_mile)
axlabels = np.arange(0, 6.5, .25)
#Plot runner density per block for each time interval
for i in [1200, 1800, 7200]:
#for i in intervals:
    fig = plt.figure()
    fig.set_size_inches(18,6)
    cols = 'Pos_' + str(i)
    timeblocks[cols][timeblocks[cols] >= 0].hist(bins=num_blocks)
    plt.xlim(0,num_blocks)
    plt.ylim(0,25)
    plt.title('Runner Density per ' + str(blocksize) + '-Foot Block of the Bridge at ' + str(racestart + dt.timedelta(seconds=i)))
    plt.xlabel('Position on Bridge (mile marker)')
    plt.ylabel('Number of Runners')
    plt.vlines(num_blocks, 0, 25, color='red')
    plt.xticks(axseries, axlabels)
    plt.show()
```

Since there is no inter-wave passing if the waves are organized optimally, it would be possible to start the wave closer together in order to end the race earlier and save money and resources.  This is significant because closing the bridge is a major undertaking that requires lots of security and logistical planning, and causes a great deal of traffic backup. This increased the total number of runners on the bridge to 14000 at the peak, but only for a very short time, and since it brought the inter-wave passing rate down to zero, it likely would not be an issue with respect to crowding.  Of course there would still be inter-wave passing in reality as runners' estimated times were not good predictors of their actual time.  This could be minimized somewhat by using runners' times from previous races to organize the waves instead of asking them to estimate.  There is a clear tradeoff between the number of runners on the bridge and total bridge closure time to consider in planning this race.  Also, the closure time could be shortened even more by starting some of the slowest runners earlier.  There would be a tradeoff in doing this though with the inter-wave passing rate, as moving any slow runners earlier increases passing.  As long as passing is minimized it should not have a significant effect on crowding - the trick is figuring out how many runners to move earlier in order to reduce the closure time while avoiding creating too much congestion due to inter-wave passing.  

So the major tradoffs and constraints to consider are: 
<p>-Keep runners on bridge below a maximum capacity
<p>-Minimize clumping - no more than 8 runners across in any spot on the bridge at once (minimum of 3 feet per runner)
<p>-Minimize inter-wave passing rate (minimum attained when runners are organized from fastest to slowest)
<p>-Minimize bridge closure time (minimum attained when runners are organized from slowest to fastest and waves are spaced as close together as possible)

I did some research on race wave organization but the only result I found was for the Paine to Pain Half Marathon race in New York (http://www.painetopain.com/2013/03/chip-timing-and-seeding-for-2013/).  The web page for this race describes the algorithm they used for wave organization.  Instead of organizing the waves directly by estimated finish times, they used an overlapping wave method where they still assigned the very fastest runners to the earliest waves since First, Second, and Third place finishers are determined from the Gun time as opposed to the Chip time (which is used to determine the rest of the awards).  They also assigned the slowest runners to the last wave.  Then they used overlapping waves to break up the large blocks of 9:00 minute mile runners - this was actually designed to prevent crowding, and even though they acknowledged that it would result in more passing, the gradual increase in pace along with the waves assured that the passing would be gradual and not cause crowding.  Finally, they enforced wave assignments by disqualifying runners who started in waves earlier than assigned (starting in later waves to run with friends was permitted).  This is an example of the algorithm they used:

Overlapping wave algorithm 
<br>Wave 1: 6:15 – 8:30  (9:00 a.m.)
<br>Wave 2: 7:30 - 9:00  (9:04 a.m.)
<br>Wave 3: 8:00 – 9:30  (9:08 a.m.)
<br>Wave 4: 8:30 – 10:00 (9:12 a.m.)
<br>Wave 5: 9:00 – 10:30  (9:16 a.m.)
<br>Wave 6: 10:00 – 12:00  (9:20 a.m.)

This made me think that for this race, organizing runners sequentially by pace in order to prevent passing might not be the best option, as putting large numbers of runners with similar paces would have much more of a negative impact with respect to crowding than would allowing for some passing.  I took a look at the time distributions in each of my "optimal" waves (the waves organized sequentially by actual finish times) and calculated the standard deviations to measure how spread out each of these hypothetical waves would be.  The earlier and later waves had relatively large standard deviations.  The middle waves had small standard deviations, and were made up with racers of almost all exactly the same pace around the median finish time.  Organizing the waves in this manner would almost certainly be detrimental to race flow, as the middle waves would be very crowded with all runners running at almost the exact same pace.

```python
for i in range(1, 11):
    xwave = results_all[results_all['Optimal_Wave'] == i]
    mintime = min(xwave['Time_Sec'])
    maxtime = max(xwave['Time_Sec'])
    wavesd = np.std(xwave['Time_Sec'])
    #Calculate hours
    mintime_h = int(mintime/3600)
    maxtime_h = int(maxtime/3600)
    #Calculates minutes
    mintime_m = int((mintime - (mintime_h*3600))/60)
    maxtime_m = int((maxtime - (maxtime_h*3600))/60)
    #Calculate seconds
    mintime_s = int(round(mintime - (mintime_h*3600) - (mintime_m*60), 0))
    maxtime_s = int(round(maxtime - (maxtime_h*3600) - (maxtime_m*60), 0))
    #Create string with statistics:
    mintime = str(mintime_h) + ':' + str(mintime_m) + ':' + str(mintime_s)
    maxtime = str(maxtime_h) + ':' + str(maxtime_m) + ':' + str(maxtime_s)
    print 'Wave:', i, mintime, ' - ', maxtime, 'SD:', wavesd
```
