I created this separate notebook to run the runner density by time model from my final project.  Here I created a plot of runner density on each 10 foot block of the bridge for every 10 minute increment of the race.  These plots show some intereting trends.  While the times from the overall race created a skewed distribution, the density plots show each wave appear to start off with a normal distribution, then spread out over time.  Toward the beginning of the race (when runner estimated times were more accurate), runners were more clumped together, but they were more evenly distributed across the bridge during the later waves.

```python
%matplotlib inline 

#Import necessary packages
import datetime as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataframes from backup CSV files
results_all = pd.read_csv('results_all8.csv')
#Convert times to pandas datetime objects
results_all['Start_Time'] = pd.to_datetime(results_all['Start_Time'])
results_all['Finish_Time'] = pd.to_datetime(results_all['Finish_Time'])
#results_all['Time_Delta'] = pd.to_timedelta(results_all['Time_Delta'])
#Correct date to be the date of the race
results_all['Start_Time'] = results_all.apply(lambda r: r['Start_Time'].replace(year = 2014, month=11, day=9), axis=1)
results_all['Finish_Time'] = results_all.apply(lambda r: r['Finish_Time'].replace(year = 2014, month=11, day=9), axis=1)
```

```python
#Set block size in feet
blocksize = 10
#Calculate number of blocks in the race course (10k = 6.2 miles)
num_blocks = np.ceil((5280*6.2)/blocksize)
#Set racestart to equal the start time of the first runner
racestart = min(results_all['Start_Time'])
#Convert start and finish times to seconds (divide by 1 second to convert to float, then convert to int)
results_all['Start_Time_Sec'] = (results_all['Start_Time'] - racestart)/dt.timedelta(seconds=1)
results_all['Start_Time_Sec'] = results_all['Start_Time_Sec'].astype('int')
results_all['Finish_Time_Sec'] = results_all['Start_Time_Sec'] + results_all['Time_Sec']
timeblocks = results_all[['Bib', 'Start_Time_Sec', 'Finish_Time_Sec', 'Time_Sec']]
#Calculate runner block pace (in blocks/second)
timeblocks['Pace'] = num_blocks/(timeblocks['Time_Sec'].astype('float'))

#Create array of times to take measurements
intervals = []
time_counter = 0
#Loop through and create a column for each 5 minute increment of the race that records the position of all runners at that time
while time_counter < 18000:
        #Use increments of 10 minutes
        time_counter += 600
        intervals.append(time_counter)
        #Form column name as Pos_[Time]
        col_name = 'Pos_' + str(time_counter)
        #Calculate block position at the time for each runner
        timeblocks[col_name] = ((time_counter - timeblocks['Start_Time_Sec'])*(timeblocks['Pace']) + 1).astype('int')
        #Set block value of runners who haven't started yet to -2
        timeblocks[col_name][timeblocks[col_name] < 1] = -2
        #Set block value of runners who finished already to -1
        timeblocks[col_name][timeblocks[col_name] > num_blocks] = -1
```

```python
blocks_per_quarter_mile = num_blocks/(6.2*4)
#Calculate x-axis labels to show mile markers instead of block number
axseries = np.arange(0, num_blocks + blocks_per_quarter_mile, blocks_per_quarter_mile)
axlabels = np.arange(0, 6.5, .25)
#Plot runner density per block for each time interval
for i in intervals:
    fig = plt.figure()
    fig.set_size_inches(18,6)
    cols = 'Pos_' + str(i)
    timeblocks[cols][timeblocks[cols] >= 0].hist(bins=num_blocks)
    plt.xlim(0,num_blocks)
    plt.ylim(0,25)
    plt.title('Runner Density per ' + str(blocksize) + '-Foot Block of the Bridge at ' + str(racestart + dt.timedelta(seconds=i)))
    plt.xlabel('Position on Bridge (mile marker)')
    plt.ylabel('Number of Runners')
    plt.vlines(num_blocks, 0, 25, color='red')
    plt.xticks(axseries, axlabels)
    plt.show()
```
